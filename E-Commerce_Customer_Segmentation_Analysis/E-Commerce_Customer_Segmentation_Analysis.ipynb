## Gender based analysis

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_amount = df.groupby('Gender')['Amount'].sum().reset_index()
```

```python
sns.barplot(x='Gender', y='Amount', data=sales_amount)
```

##### Above analysis stats that Female buyers are more and they spent more amount than male.

## Age Group wise analysis

```python
ax = sns.countplot(x='Age Group', hue= 'Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Age_sale = df.groupby('Age Group')['Amount'].sum().reset_index().sort_values('Amount',ascending=False)
```

```python
sns.barplot(x='Age Group' , y='Amount', data=Age_sale)
```

##### Above analysis depicts that most of the buyers are from the age 26-35 yrs female

### State based analysis

```python
state_orders = df.groupby('State')['Orders'].count().reset_index().sort_values('Orders',ascending=False).head(10)
plt.figure(figsize=(25,10))
sns.barplot(x='State', y='Orders', data=state_orders)
```

```python
state_amount = df.groupby('State')['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(10)
plt.figure(figsize=(25,10))
sns.barplot(x='State', y='Amount', data=state_amount)
```

##### Above analysis stats that Utter Pradesh has the highest orders followed by Maharastra, Karnataka and same tren is observed in terms of spending. However, there is unusual trend is been shown for Kerala.

### Marital Status analysis

```python
ax = sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
maritalstatus_data = df.groupby('Marital_Status')['Amount'].sum().reset_index().sort_values('Amount',ascending=False)
```

```python
sns.barplot(x='Marital_Status' , y='Amount', data=maritalstatus_data)
```

##### From above graps we can see that there are more number of married people and they spend more money than unmarried people.

### Occupation analysis

```python
plt.figure(figsize=(25,10))
ax = sns.countplot(x='Occupation', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_data = df.groupby('Occupation')['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(10)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='Occupation' , y='Amount', data=occupation_data)
```

##### Above analysis depicts that IT sector is on the top in terms of spending and number of orders followed by Healthcare and Aviation

### Product Category Analysis

```python
plt.figure(figsize=(25,10))
ax = sns.countplot(x='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(25,10))
product_data = df.groupby('Product_Category')['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(10)
sns.barplot(x='Product_Category' , y='Amount', data=product_data)
```

##### Above graphs shows that most sold products are from Clothing & apparel but Food category is generating highest revenue
